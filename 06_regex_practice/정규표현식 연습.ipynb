```python
# 부분 팰린드롬, i,i+1부터 시작하는 중 제일 긴 것 찾기

data = 'abaabbaac'
# data = 'ewqpbewqbfjabcdefedcbaienqnfkndkl'

# 회문의 종류
# 1. 짝수(펠린드)롬:  aa baab -> i, i+1  -> left, right로 퍼져나가는데,    옮긴 i>=0, 옮긴i+1 <= n-1
# 2. 홀수롬 : aca   bacab  -> i, i+2 로 퍼져나가기
# * 기준을 맨 앞인 i로 잡아서, 인덱스 검사를 ii+1, ii+2로 편하게 한다.

# 2. 애초에 팰린드롬이면, 탐색할 필요가 없다.
if len(data) <2 or data == data[::-1]: 
    print(data)
    exit()

# 3. 각 i마다, 짝수룸(i,i+1) 홀수롬(i,i+2)을 직접 [시작2점으로 대입]해주면 (현재항) -> left, right의 시작점으로서, 가장 긴 팰린드롬을 추출
# * 인수 : i,i+1  or i, i+2 -> 각 i는 변수, i+1, i+2는 i에 종속이나..각각이 다른 경우니 변수
def palindrome(left, right):
    # *left, right가  i,i+1(짝수롬) 일지,  i,i+2(홀수롬)일지는 모른다.
    # left, right의 조건으로 먼저 필터링한다. 
    # if left>=0 and right < len(data):
    #     if data[left] == data[right]:
    #         # 팰린드롬이면, 한칸씩 이동해준다. -> 체크는???  안해주고.. 옮겨간 left, right(index) -> [인덱스만 있으면 길이]를 알 수 있음.
    #         left-=1
    #         right+=1
    #         # 이동후 또 검사하고... 반복해야하니 while 

    while left>= 0 and right<len(data) and data[left] == data[right]:
        left -=1
        right +=1


    if left <0:
        left+=1
    if right >= len(data):
        right-=1
    
    return data[left:right+1]


# 4. 이제 매 i를 돌면서 i,i+1 및 i,i+2를 시작점으로 하는 팰린드롬 추출 -> 전체에서 가장 긴 길이최대값 추출하기
max_len_pal = ''
for i in range(len(data)):
    # print(palindrome(i, i+1)) # 짝수롬
    # print(palindrome(i, i+2)) # 홀수롬
    #* max(, key=len) : 인자의 val값이 아닌, key = len 길이를 기준으로 최대값을 뽑는다.
    max_len_pal = max(max_len_pal, palindrome(i,i+1), palindrome(i,i+2), key=len)

max_len_pal
```

```python
# 빈도 dict -> ranking 1부터 dict로 변경
# 정렬 -> 정렬된 순서 -> enumerate로 활용하기.
sentences = [['barber', 'person'], ['barber', 'good', 'person'], ['barber', 'huge', 'person'], ['knew', 'secret'], ['secret', 'kept', 'huge', 'secret'], ['huge', 'secret'], ['barber', 'kept', 'word'], ['barber', 'kept', 'word'], ['barber', 'kept', 'secret'], ['keeping', 'keeping', 'huge', 'secret', 'driving', 'barber', 'crazy'], ['barber', 'went', 'huge', 'mountain']]

from itertools import chain

words = list(chain(*sentences))
vocab = Counter(words)

vocab.most_common(5) # 튜플lst의  [빈도순으로 정렬된 순서]를 이용 -> 돌면서 enumerate로 [index]를 활용

rank_dict = {}
for num, (word, cnt) in enumerate(vocab.most_common(5), start=1) :
    rank_dict[word] = num 

rank_dict



```

```python
def isPalindrome(arr):
    for i in range(len(arr)//2):
        # 같으면 ==회문이면, 건너띄다가
        if not (arr[i] == arr[-(i+1)]):
            print("회문이 아닙니다.")
            break
    else:
        print("회문입니다.")

isPalindrome('aabaaa')
```

```python
# 복습1. x를 y번 누적:  y번 반복하는  ---> O(y)
def power(x,y):
    total  = 1 
    for _ in range(y):
        total *= x 
    
    return total

power(2,10)
```

```python
# 복습2. 재귀인 f(y) = f(y-1) * x로 바꿔도 ---> O(y) 
def power(x,y):
    # base
    if y == 0:
        return 1 

    return power(x,y-1)  * x
    
power(2,10)
```

```python
# 복습3. O(y)를  O(lg y)로 줄이는 재귀 + 부분문제 divide가 [ y//2 ]가 적용가능한지 보기
# - y//2가 똑같은 것이 등장한다?  y-1,y-2는 tabulation -> 공간최적화 가능  y//2 등 팍팍주는 것은 top-down으로 -> memoization 사용
# - y//2: 절반씩 줄어들 때, //2에서 2의배수가 아니라면(당연히 나누어지는 것을 int목적으로 나누기대신 사용하는게 아니라면), 홀수인지 짝수인지 판단필수!!
# -- 6//2 = 3,  5//2 = 2 -> 2 * 2 -> 1더 필요.. -> x를 그냥 한번 더 곱해줘야한다.
def power(x,y):
    # base
    if y == 0:
        return 1 

    # recursive
    if y%2==0:
        return power(x, y//2)  * power(x, y//2)
    else:
        return x * power(x, y//2)  * power(x, y//2)

    
    
power(2,10)
```

```python
import bisect

def bs_by_bisect(sorted_list, val):
    insert_point = bisect.bisect(sorted_list, val)
    return insert_point-1 if sorted_list[insert_point-1] == val else -1, f"insert to {insert_point}"


bs_by_bisect(mylist, 8)
```

```python
def quick_sort(array):
    if len(array) <= 1:
        return array 

    pivot = array[0]
    tail = array[1:]

    left_side = [ x for x in tail if x <= pivot]
    right_side = [ x for x in tail if x > pivot]

    return quick_sort(left_side) + [pivot] + quick_sort(right_side)

quick_sort([5, 7, 9, 0, 3, 1, 6, 2, 4, 8])
```
